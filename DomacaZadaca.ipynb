```
import pandas as pd
from sklearn import preprocessing

x = bc.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```
xtrain = df.iloc[0:3918, 0:11]
```

```
ytrain = df.iloc[0:3918, 11]
```

```
xtrain.head()
```

```
ytrain.head()
```

```
xtest = df.iloc[3918:, 0:11]
ytest = bc.iloc[3918:,11]
```

```
xtest
```

```
from keras.models import Sequential
from keras.layers import Dense
```

```
model = Sequential()

model.add(Dense(30, activation = 'softplus', input_dim = 11))
model.add(Dense(15, activation = 'softplus', input_dim = 11))  #broj neurona ulaznog sloja
model.add(Dense(units = 1, activation = 'sigmoid' )) #izlazni sloj
```

```
model.summary()
```

```
model.compile(optimizer= 'Adam', loss= 'mean_squared_error')
```

```
model.fit(xtrain,ytrain, batch_size= 3, epochs= 10)
```

```
y_pred = model.predict(xtest)
```

```
y_pred
```

```
import numpy
y_pred_results = [int(numpy.round(x*10)) for x in y_pred]

y_pred_results
```

```
total = 0
correct = 0

for i in range(len(y_pred_results)):
  total = total + 1
  if(ytest.iloc[i] == y_pred_results[i]):
    correct = correct + 1
```
